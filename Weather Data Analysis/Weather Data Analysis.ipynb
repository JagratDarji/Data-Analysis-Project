Step 1 : import library

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

Step 2 : Load Dataset

```python
df = pd.read_csv(r"C:\Users\Jagrat Darji\Desktop\Data Analytics\Weather Data Analysis\Dataset\weatherHistory.csv")
```

Step 3 : Convert date

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
```

Step 4 : clean Data 

```python
df.dropna(subset=['Precip Type'], inplace=True)
```

Step 5 : Summary

```python
print(df.describe())
```

Step 6 : Plot 1: Temperature over time

```python
plt.figure(figsize=(12, 5))
sample_df = df.sample(10000, random_state=42)
plt.plot(sample_df['Formatted Date'], sample_df['Temperature (C)'], color='orange', linewidth=0.5)
plt.title("Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.grid()
plt.show()
```

Step 7 :  Plot 2: Temperature distribution

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Temperature (C)'], kde=True, color='red')
plt.title("Distribution of Temperature")
plt.xlabel("Temperature (C)")
plt.show()
```

Step 8 : Plot 3: Boxplot of temperature by Precip Type

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Precip Type', y='Temperature (C)', data=sample_df)
plt.title("Temperature by Precip Type")
plt.show()
```

Step 9 : Plot 4: Humidity vs Temperature

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=sample_df, x='Temperature (C)', y='Humidity', hue='Precip Type', alpha=0.6)
plt.title("Humidity vs Temperature")
plt.show()
```

Step 10 :  Plot 5: Correlation heatmap

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
